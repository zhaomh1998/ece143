```python
import preprocessing
```

```python
household_characteriscits = preprocessing.INECDataSet('household_characteristics')
print(household_characteriscits.get_dataset_description())
hh_char = household_characteriscits.get_dataset()
hh_char.head()
```

```python
year_dict = preprocessing.INECDataSet('edu').get_dataset()
year_dict
```

```python
import pandas as pd
edu = pd.DataFrame()
for year, df in year_dict.items():
    df = df.drop('Unnamed: 0', axis=1)
    df = df.dropna()
    df = df.rename(dict(zip(list(df.columns), ['region', '5-24 total', '5-24 attend%', '5-24 not attend%',
                                              '5-12 total', '5-12 attend%', '5-12 not attend%',
                                              '13-17 total', '13-17 attend%', '13-17 not attend%',
                                              '18-24 total', '18-24 attend%', '18-24 not attend%'])), axis=1)
    df['year'] = year
    edu = edu.append(df)
```
