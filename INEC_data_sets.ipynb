```python
import os
import pandas as pd
edu_lvl = pd.read_excel(pd.ExcelFile(os.getcwd() + '/datasets/edu_2017.xlsx'), 1)
edu_lvl = edu_lvl.drop('Unnamed: 0', axis=1)
descrip = preprocessing.string(edu_lvl.iloc[:3,0].str.cat(sep='\n')).translate()
print(descrip)
```

```python
edu_lvl
```

```python
edu_lvl_cols = ['region', 'total', 'no education', 'incomplete primary', 'complete primary',
                'secondary incomplete', 'secondary complete', 'technical secondary incomplete', 'technical secondary complete',
                'undergraduate', 'postgraduate', 'no response']
edu_lvl = edu_lvl.rename(dict(zip(edu_lvl.columns, edu_lvl_cols)), axis=1).dropna(subset=['region'])
edu_lvl['male'] = (edu_lvl['region'] == 'Hombres').astype(int)
edu_lvl['total'] = ((edu_lvl['region'] != 'Hombres') & (edu_lvl['region'] != 'Mujeres')).astype(int)
```

```python
for index, item in edu_lvl['region'].iteritems():
    if item == 'Hombres':
        edu_lvl.loc[index,'region'] = edu_lvl['region'][index-1]
    elif item == 'Mujeres':
        edu_lvl.loc[index,'region'] = edu_lvl['region'][index-2]
edu_lvl = edu_lvl.dropna()
```
